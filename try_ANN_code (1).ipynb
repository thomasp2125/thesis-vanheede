```python
def mean_absolute_percentage_error(actual, pred):
    actual = np.array(actual)
    pred   = np.array(pred)
    # Calculate MAPE (%) – add a small epsilon to avoid division by zero if needed
    return np.mean(np.abs((actual - pred) / (actual + 1e-8))) * 100

def directional_mean_accuracy(actual, pred):
    actual = np.array(actual)
    pred   = np.array(pred)
    # Calculate the sign of changes for actual and predicted
    actual_direction = np.sign(actual[1:] - actual[:-1])
    pred_direction   = np.sign(pred[1:] - pred[:-1])
    # Compare directions and compute accuracy as a percentage
    correct_direction = (actual_direction == pred_direction)
    return np.mean(correct_direction) * 100
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# --- CONFIGURATION ---
LAG = 144
OUT_HORIZON = 24
EPOCHS = 100
BATCH_SIZE = 64
PATIENCE = 15

# --- 1. Data Preparation ---
df = df_csv.copy()
df.index = pd.to_datetime(df.index)
series = df["Load"]

val_start = pd.to_datetime("2024-07-29")
test_start = pd.to_datetime("2024-10-07")

train_series = series.loc[: val_start - pd.Timedelta(hours=1)]
val_series = series.loc[val_start: test_start - pd.Timedelta(hours=1)]
test_series = series.loc[test_start:]

def normalize_series(s, ref=None):
    mu, sigma = ref if ref else (s.mean(), s.std())
    return (s - mu) / sigma, (mu, sigma)

def rescale_series(s, params):
    mu, sigma = params
    return s * sigma + mu

norm_train, norm_params = normalize_series(train_series)
norm_val, _ = normalize_series(val_series, norm_params)
norm_test, _ = normalize_series(test_series, norm_params)

def create_day2_data(series, lag=144, keep=24, offset=24):
    X, Y = [], []
    values = series.dropna().values
    for i in range(len(values) - lag - offset - keep + 1):
        X.append(values[i:i + lag])
        Y.append(values[i + lag + offset:i + lag + offset + keep])
    return np.array(X)[..., np.newaxis], np.array(Y)

X_train, y_train = create_day2_data(norm_train, lag=LAG, keep=OUT_HORIZON, offset=24)
X_val, y_val = create_day2_data(norm_val, lag=LAG, keep=OUT_HORIZON, offset=24)

def build_lstm_model(units, n_layers=1, dropout=0.0):
    model = Sequential()
    model.add(Input(shape=(LAG, 1)))
    for i in range(n_layers):
        return_seq = (i < n_layers - 1)
        model.add(LSTM(units, return_sequences=return_seq))
        if dropout > 0:
            model.add(Dropout(dropout))
    model.add(Dense(OUT_HORIZON))
    model.compile(optimizer='adam', loss='mse')
    return model

def forecast_day2_direct(series, model, lag=144, keep=24, offset=24):
    forecasts, timestamps = [], []
    for start in pd.date_range(series.index[0] + pd.Timedelta(hours=lag + offset),
                               series.index[-1] - pd.Timedelta(hours=keep) + pd.Timedelta(hours=1),
                               freq='D'):
        end = start - pd.Timedelta(hours=offset + 1)
        start_window = end - pd.Timedelta(hours=lag - 1)
        if start_window not in series.index or end not in series.index:
            continue
        window = series.loc[start_window:end].values
        if len(window) != lag or np.isnan(window).any():
            continue
        window = window.reshape((1, lag, 1))
        preds = model.predict(window, verbose=0).flatten()
        forecast_time = start
        index = pd.date_range(forecast_time, periods=keep, freq='h')
        forecasts.append(preds)
        timestamps.append(index)
    return pd.Series(np.concatenate(forecasts), index=np.concatenate(timestamps))

# --- 2. Grid Search ---
configs = [
    {"units": 32, "n_layers": 1, "dropout": 0.0},
    {"units": 64, "n_layers": 1, "dropout": 0.0},
    {"units": 128, "n_layers": 1, "dropout": 0.1},
    {"units": 64, "n_layers": 2, "dropout": 0.1},
    {"units": 128, "n_layers": 2, "dropout": 0.2},
]

results = []

for cfg in configs:
    print(f"Training model: {cfg}")
    model = build_lstm_model(**cfg)
    es = EarlyStopping(monitor="val_loss", patience=PATIENCE, restore_best_weights=True)
    model.fit(X_train, y_train,
              validation_data=(X_val, y_val),
              epochs=EPOCHS,
              batch_size=BATCH_SIZE,
              callbacks=[es],
              verbose=0)

    pred_series = forecast_day2_direct(norm_test, model, lag=LAG, keep=OUT_HORIZON, offset=24)
    pred_series = rescale_series(pred_series, norm_params)
    y_true = series.loc[pred_series.index].values
    y_pred = pred_series.values
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    results.append((cfg, rmse))

    # Plot
    plt.figure(figsize=(12, 4))
    plt.plot(series.loc[pred_series.index[0]:pred_series.index[-1]], label='True Load')
    plt.plot(pred_series, label='Forecast', linestyle='--')
    plt.title(f"Config: {cfg} — RMSE: {rmse:.2f}")
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 3. Summary ---
results.sort(key=lambda x: x[1])
for cfg, rmse in results:
    print(f"Config {cfg} → RMSE = {rmse:.2f}")
```
